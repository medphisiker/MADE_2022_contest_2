```python
import numpy as np
import pandas as pd
```

```python
train_true = pd.read_csv('data/train.csv')
```

```python
train_true
```

```python
train_pred = pd.read_csv('data/algo_train_pred.csv')
```

```python
train_pred
```

```python
y_true = train_true.label.values
y_pred = train_pred.label.values
```

```python
# Введем метрики оценки качества модели
def mse(y_true, y_pred):
    return ((y_true - y_pred)**2).mean()

def mae(y_true, y_pred):
    return (np.abs(y_true - y_pred)).mean()
```

```python
# Оценим нашу модель на валидационной выборке:
print(f"MSE = {mse(y_true, y_pred)}")
print(f"MAE = {mae(y_true, y_pred)}")
```
